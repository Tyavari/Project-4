```python
# import dependencies 
import pandas as pd

```

```python
#load csv
nyc_sales = "NYC_sales.csv"

nyc_sales_df = pd.read_csv(nyc_sales)

nyc_sales_df.head()
```

```python
# Data cleaning 
# no easement data
# brough block and lot are redundent as we have adress and neighborhood
nyc_sales_df2 = nyc_sales_df.drop(["BOROUGH","BLOCK", "LOT", "EASE-MENT","ADDRESS", "APARTMENT NUMBER"], axis=1)
nyc_sales_df2.head()
```

```python

nyc_sales_df2.dropna(subset=['SALE PRICE'])
```

```python
#Cast Sale date to a sting
nyc_sales_df2['SALE DATE'] = nyc_sales_df2['SALE DATE'].astype(str)
```

```python
# Split sales date
df_temp = nyc_sales_df2['SALE DATE'].str.split('/', expand=True)
print(df_temp)
```

```python
#temporary data base to split
df_temp = df_temp.rename(columns = {0: "MONTH",2: "YEAR"})
df_temp
```

```python
df_temp2 = df_temp["YEAR"].str.split(' ', expand=True)
df_temp2
```

```python
df_temp3 = df_temp2.drop([1], axis = 1)
df_temp3
```

```python
# create/concatenate new column
df_temp3 = df_temp3.rename(columns = {0: "YEAR"})

df_temp3["MONTH"]= df_temp["MONTH"]

df_temp3["SALE PERIOD"] = df_temp3["MONTH"]+"-"+df_temp3["YEAR"]

df_temp3= df_temp3.drop(["YEAR", "MONTH"], axis = 1)
df_temp3
```

```python
#Merge back on to nyc_sales_df2
nyc_sales_df2 = pd.merge(nyc_sales_df2, df_temp3, left_index=True, right_index=True)
```

```python
#drop original Sale date column
nyc_sales_df2 = nyc_sales_df2.drop(["SALE DATE"], axis=1)
```

```python
# Further cleaning
nyc_sales_df2 = nyc_sales_df2.drop(["Unnamed: 0", "TAX CLASS AT PRESENT","BUILDING CLASS AT PRESENT"], axis=1)
```

```python
nyc_sales_df2.dropna(inplace = True)
```

```python
#Rename column 
nyc_sales_df2= nyc_sales_df2.rename(columns = {"SALE PRICE": "SALE_PRICE"})
```

```python
nyc_sales_df2
```

```python
# remove Blanks
nyc_sales_df2 = nyc_sales_df2[nyc_sales_df2.SALE_PRICE !=" "]
```

```python
# Export df to new csv
nyc_sales_df2.to_csv("NYC_Sales_Clean.csv", index=False)
```
